Las columnas "PM2.5", "PM10", "NO2", "O3" y "SO2" hacen referencia a diferentes tipos de contaminantes atmosféricos.

1. **PM2.5**:
   - Son partículas suspendidas en el aire con un diámetro de 2.5 micrómetros o menos. Pueden incluir polvo, hollín, cenizas y gotas líquidas. Son lo suficientemente pequeñas como para ser inhaladas y pueden tener efectos perjudiciales en la salud humana.

2. **PM10**:
   - Son partículas suspendidas en el aire con un diámetro de 10 micrómetros o menos. Esto incluye partículas de polvo, polen, esporas de moho y cenizas. También pueden tener impactos en la salud, especialmente si son inhaladas.

3. **NO2** (Dióxido de nitrógeno):
   - Es un gas de color rojizo y olor penetrante. Es un subproducto de la combustión de combustibles fósiles, como los vehículos de motor y las centrales eléctricas. La exposición prolongada a niveles altos de NO2 puede tener efectos negativos en la salud respiratoria.

4. **O3** (Ozono):
   - Aunque el ozono es beneficioso en la estratosfera (donde forma la capa de ozono que protege la vida en la Tierra de la radiación ultravioleta), a nivel del suelo puede ser perjudicial. Es un componente importante de la contaminación del aire y puede causar problemas respiratorios.

5. **SO2** (Dióxido de azufre):
   - Es un gas incoloro con un olor penetrante y desagradable. Se libera principalmente durante la combustión de carbón que contiene azufre y durante procesos industriales. La exposición a altos niveles de SO2 puede causar problemas respiratorios y otros efectos adversos para la salud.

Estos contaminantes atmosféricos son monitoreados porque tienen un impacto significativo en la calidad del aire y, por lo tanto, en la salud de las personas y el medio ambiente. Los niveles elevados de estos contaminantes pueden contribuir a problemas de salud como enfermedades respiratorias, cardiovasculares y otros efectos adversos.

1. **Distribución de la población por edad**:
   - Representar la distribución de la población por grupos de edad ('0-15 años', '16-64 años', '65 o más años') en diferentes municipios o distritos. Gráfico de barras apiladas:

```python
# Agrupar el DataFrame por 'Municipio'
df_grouped = df.groupby('Distrito').sum()
# Obtener los nombres de los municipios
distritos = df_grouped.index
# Obtener los valores de las edades
edad_0_15 = df_grouped['0-15 años']
edad_16_64 = df_grouped['16-64 años']
edad_65_mas = df_grouped['65 o más años']
# Definir la posición de las barras
bar_width = 0.35
bar_positions = np.arange(len(distritos))
# Crear el gráfico de barras apiladas
plt.figure(figsize=(10, 6))

plt.bar(bar_positions, edad_0_15, width=bar_width, label='0-15 años', color='b')
plt.bar(bar_positions, edad_16_64, width=bar_width, bottom=edad_0_15, label='16-64 años', color='g')
plt.bar(bar_positions, edad_65_mas, width=bar_width, bottom=edad_0_15+edad_16_64, label='65 o más años', color='r')

# Añadir etiquetas y título
plt.xlabel('Distrito')
plt.ylabel('Población')
plt.title('Distribución de la Población por Edad en Diferentes Distritos')
plt.xticks(bar_positions, distritos, rotation=45)
plt.legend()

# Guarda el gráfico como una imagen JPG
plt.savefig('Distribución_población.jpg', format='jpg', dpi=400)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
#Coclusion en el distrito de salamanca es donde mas niños menores de 15 años hay, y en carabanchel, chamberí, ciudad lineal y latina donde mas ancianos:
#para esto ultimo veremos la correlación entre los datos de la columna 16-64 años y la población:
# Selecciona las columnas relevantes
subset_df = df[['16-64 años', 'Población']]

# Calcula la matriz de correlación
correlation_matrix = subset_df.corr()

# Obtén el coeficiente de correlación entre '16-64 años' y 'Distrito'
correlation_value = correlation_matrix.loc['16-64 años', 'Población']

print(f"Coeficiente de correlación: {correlation_value}")

#Como vemos hay una alta correlación entre la poblacion y la edad adulta, a mas poblacion más edad adulta, esto explica que en la grafica anterior las barras verdes sean de tamaño parecido
```
